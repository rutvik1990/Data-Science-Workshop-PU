```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
%matplotlib inline

ranking = np.random.normal(size=28)

colors = np.array(["#FFA07A", "#E9967A", "#808000", "#FFBFBF", "#ff3399", "#3333ff","#99e6ff","#FFA07A", "#E9967A", "#808000", "#FFBFBF", "#ff3399", "#3333ff","#99e6ff","#FFA07A", "#E9967A", "#808000", "#FFBFBF", "#ff3399", "#3333ff","#99e6ff","#FFA07A", "#E9967A", "#808000", "#FFBFBF", "#ff3399", "#3333ff","#99e6ff"])

fig, ax = plt.subplots(figsize=(18,10))

plt.scatter(corona_dataset_csv["Total deaths/million"],corona_dataset_csv["Average vitamin D levels (nmol/L)"], s = 1000,c =colors,alpha = 0.7)

m, b = np.polyfit(corona_dataset_csv["Total deaths/million"], corona_dataset_csv["Average vitamin D levels (nmol/L)"], 1)

plt.title("Total deaths/million vs Average vitamin D levels (nmol/L)")

#list1 = list(range(20,200,5))

#plt.xticks(list1)

plt.xlabel("Total deaths/million")
plt.ylabel("Average vitamin D levels (nmol/L)")

plt.plot(corona_dataset_csv["Total deaths/million"], (m*corona_dataset_csv["Total deaths/million"]) + b)

for i, txt in enumerate(corona_dataset_csv.index):
    plt.annotate(txt, (corona_dataset_csv["Total deaths/million"][i],corona_dataset_csv["Average vitamin D levels (nmol/L)"][i]), fontsize = 12)
print(corona_dataset_csv.shape)
plt.savefig("deaths_vitamind.png", dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', format=None,
        transparent=True, bbox_inches=None, pad_inches=0.1)
```
